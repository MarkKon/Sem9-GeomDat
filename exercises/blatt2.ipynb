18. Illustriere mit Python eine Kurve, deren Kr√ºmmung linear zunimmt, wenn sie mit konstanter Geschwindigkeit zunimmt:

```python
%config InlineBackend.figure_format='svg'

d = .01
length = 10

def interpol_circle(x_1,x_2, r):
  dx = x_2 - x_1
  theta = np.arctan2(dx[1], dx[0])
  phi = np.arcsin(d/(2*r))
  return (x_2 + d* np.array([np.cos(theta+2*phi), np.sin(theta+2*phi)]), phi, theta)

a = [(np.array([0,0]),0,0), (np.array([d,0]), 0,0)]
for i in range(1,int(length/d)):
  a.append(interpol_circle(a[-2][0], a[-1][0],1/(d*i)))

coords, phi, theta = zip(*a)
x,y = zip(*coords)
fig, ax = plt.subplots(figsize=(6, 6))
ax = cartesian_coordinates(ax, xlim=(0, 2), ylim=(0,2), ticks_frequency=1)
plt.plot(x,y, lw= 1)
plt.show()
```
